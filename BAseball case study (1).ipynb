```python
#removing outliers using zscore, zscore is one of the libraries inside scipy.stats by which we can remove the outliers
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
#threshold for zscore-- above three are considered as outliers
threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
df_new
```

## Splitting our data into training and testing

```python
y=df_new.iloc[:,0]
y
```

```python
x=df_new.iloc[:,1:]
x
```

```python
x.shape
```

```python
y.shape
```

## Scaling our data


```python
#using standard scaler
from sklearn.preprocessing import StandardScaler

st=StandardScaler()
x=st.fit_transform(x)
x
```

```python
xf=pd.DataFrame(data=x)
xf

#this entire data is scaled data
```

```python
#now we make the train test split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20, random_state=45)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train, y_train)
```

98% our model has learnt through data

```python
#predict the value
pred=lm.predict(x_test)
print("Predicted result price:", pred)
print("Actual Price", y_test)
```

```python
#performing model validation
#these are the metrices which are giving me the performance of the module
print('error:')
print('mean absolute error:', mean_absolute_error(y_test, pred))
print('mean squared error:', mean_squared_error(y_test, pred))
print('root mean squared error:', np.sqrt(mean_squared_error(y_test, pred)))
```

```python
#r2 score -->coefficient of determination
#is the proportation of the variance in the dependent variable that is predictable from the independent variable(s).
#i.e. change coming in y whenever x is being changed

from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```
